```python
#User Login Check
print("Welcome to Login System")
correct_username = "admin"
correct_password = "1234"
username = input("Enter Username: ")
password = input("Enter Password: ")
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60, 11]
pass_count = 0
fail_count = 0
for m in marks:
    if m >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
#Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    name = name.strip()
    name = name.lower()
    cleaned_names.append(name)
print("Cleaned Names List:")
print(cleaned_names)
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)
    if length > 10:
        print("Status: Long Message")
    else:
        print("Status: Normal Message")
    print()
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR","ERROR"]
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count = error_count + 1
print("Total ERROR messages:", error_count)
```
